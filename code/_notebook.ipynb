```python
from functions.Miscellaneous import load_track_from_db, angle_between
import matplotlib.pyplot as plt
import numpy as np

track = load_track_from_db("../tracks/0-2500-oil_04_08_2025/top (0)_1/1/")
def lex_less(a, b):
    for ai, bi in zip(a, b):
        if ai < bi:
            return True
        elif ai > bi:
            return False
    return False

shape_evolution = np.zeros((len(track), len(track[0]["cell"]["shape"]), 2))

for p, pos in enumerate(track[60:]):
    shape = pos["cell"]["shape"]
    
    foward_dir = shape[1, :] - shape[0, :]
    perpendicular = np.array([foward_dir[1], -foward_dir[0]])

    vectors = [pt - shape[0, :] for pt in shape[1:]]
    closest_vector = min(vectors, key=lambda x: angle_between(x, perpendicular))

    if np.rad2deg(angle_between(closest_vector, perpendicular)) < 20:
        shape = np.vstack([shape[0], shape[1:][::-1]])
    shape_evolution[p] = shape
    # plt.scatter(shape[:, 1], shape[:, 0], c=np.linspace(0, 1, len(shape)), cmap="viridis")
    # # plt.gca().invert_yaxis()
    # plt.axis("equal")
    # plt.title(p + 60)
    # plt.show()

shape_evolution.shape
    
```

```python
import numpy as np
from pathlib import Path
import imageio.v2 as imageio
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm

tracks = np.load("../masks/0-500-oil_05_08_2025/Pos1_1/tracks.pkl", allow_pickle=True)
max_frame = max([pos["frame"] for track in tracks for pos in track])
tif_files = sorted(Path("../data/0-500-oil_05_08_2025/Pos1_1/Default/").glob("*.tif"))
print(tif_files)
background = True

images = []
colors = [plt.cm.tab20(i % 20) for i in range(len(tracks))]
for i in tqdm(range(max_frame)):
    fig, ax = plt.subplots(figsize=(6,6), dpi=300)
    for t, track in enumerate(tracks):
        color = colors[t]
        centroids = []
        last_outline = []
        for pos in track:
            frame = pos["frame"]
            if frame <= i:
                centroids.append(pos["cell"]["centroid"])
                last_outline = pos["cell"]["shape"] + pos["cell"]["centroid"]
            else:
                break
        if len(centroids) == len(track) or not len(centroids):
            continue
        
        centroids = np.array(centroids)
        ax.plot(centroids[:, 1], centroids[:, 0], color=color)
        ax.scatter(last_outline[:, 1], last_outline[:, 0], s=0.1, color=color)
    
    if background:
        background_img = imageio.imread(tif_files[i])
        ax.imshow(
            background_img,
            cmap='gray',
            origin='lower',
            vmin=850,
            vmax=2000
        )
    ax.set_axis_off()
    ax.invert_yaxis()
    
    fig.canvas.draw()
    width, height = fig.canvas.get_width_height()
    image = np.frombuffer(fig.canvas.buffer_rgba(), dtype=np.uint8)
    image = image.reshape((height, width, 4)) 
    image = image[:,:,:3] 
    images.append(image)
    plt.close(fig)

output_file = Path("../visualizations/") / f"aaaa.gif"
imageio.mimsave(output_file, images)
```
