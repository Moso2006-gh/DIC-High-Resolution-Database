# This notebook is to showcase the plotting functions

All functions have the following three parameters that can be added at the end:
- dpi (int): Control the quality of the image
- frame (int): Get only one frame instead of all the gif
- cap (int): Limit the max amount of frames you want (good for quick testing)

### Export individual frames
Each gif function returns the image array which can be exported. There is also a decorator you can use:

```python
from functions.Plotting import to_individual_frames, create_mask_gif

#To export individual frame use the decorator in this way (this is an example:

#Turn this
create_mask_gif(...)
#to this
to_individual_frames(create_mask_gif)(...)
```

## Individual Track visualization

Visualize one full track of DIC position.

```python
import copy
import numpy as np
from pathlib import Path
from functions.Plotting import create_track_gif_from_raw_tiffs

#Write here the path to the tracks file e.g "../masks/0-500-oil_05_08_2025/Pos1_1/tracks.pkl"
path_to_tracks_file = "../masks/0-500-oil_05_08_2025/Pos1_1/tracks.pkl"
n = 0 # Specify the track you are interested in


tracks = np.load(path_to_tracks_file, allow_pickle=True)
root_data = "../data/"
root_masks = "../masks/"
path_to_tifs = root_data / Path(path_to_tracks_file).parent.relative_to(root_masks) / "Default"

file_output_name = str(Path(path_to_tracks_file).parent).split("\\masks\\")[1].replace("\\", "--") + f"--track_{n}"
images = create_track_gif_from_raw_tiffs(copy.deepcopy(tracks[n]), file_output_name, "../visualizations/", background=True, path_to_tifs=path_to_tifs, 
                                flip=True, # NOTE you might need to set this variable to false or true depending on the output
                                dpi=150)
```

The following shows an example of the first frame you would get if dpi set to a low quality of 100

```python
images = create_track_gif_from_raw_tiffs(copy.deepcopy(tracks[n]), file_output_name, "../visualizations/", background=True, path_to_tifs=path_to_tifs,
                                flip=False, # NOTE you might need to set this variable to false or true depending on the output
                                dpi=100, frame=1)
```
